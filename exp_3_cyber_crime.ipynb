```python
# Sample email details

sender = input("Enter Sender Email: ")
subject = input("Enter Subject: ")
body = input("Enter Email Body: ")

# Suspicious indicators
keywords = [
    "urgent",
    "verify",
    "password",
    "bank",
    "click",
    "login",
    "winner",
    "free",
    "gift",
    "account suspended"
]

score = 0

# Check sender
if not sender.endswith(".com"):
    score += 1

# Check subject
for word in keywords:
    if word.lower() in subject.lower():
        score += 1

# Check body
for word in keywords:
    if word.lower() in body.lower():
        score += 1

print("\n------ Result ------")

if score >= 3:
    print("Phishing Email Detected")
else:
    print("Looks Like a Legitimate Email")

print("Risk Score:", score)
```
