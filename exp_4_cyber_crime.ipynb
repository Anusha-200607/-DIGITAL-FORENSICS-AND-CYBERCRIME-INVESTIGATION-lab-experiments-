```python
import socket

domain = input("Enter Domain Name: ")

try:
    ip = socket.gethostbyname(domain)

    print("\nDNS Lookup Result")
    print("-----------------------")
    print("Domain :", domain)
    print("IP Address :", ip)

except socket.gaierror:
    print("Invalid Domain Name")
```
